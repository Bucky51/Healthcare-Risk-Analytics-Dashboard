```python
import pandas as pd
import openpyxl as op
```

```python
data = pd.read_excel("..\\data\\MedicalData.xlsx")
# shape of the data
print(data.shape)

# columns in the data
print(data.columns)

# first 5 rows of the data
print(data.head(5))
```

```python
# data types and missing values
data.info()
data.isnull().sum()

# describe numeric columns
print(data.describe())

# diagnosis and hospital valuse_counts
print(data["Diagnosis"].value_counts())
print(data["Hospital"].value_counts())
```

```python
print("=== BLOOD_PRESSURE SAMPLES ===")
print(data['Blood_Pressure'].head(10).tolist())
print("\nUnique formats:", data['Blood_Pressure'].nunique(), "different values")

print("\n=== BMI EXTREMES ===")
print("Min:", data['BMI'].min(), "| Max:", data['BMI'].max())
print(data['BMI'].describe())

print("\n=== VISIT_DATE SAMPLES ===")
print(data['Visit_Date'].head(3).tolist())
```

```python
# split Blood_Pressure into Systolic and Diastolic
data[['Systolic','Diastolic']] = data['Blood_Pressure'].str.split('/',expand=True)

# convert Systolic and Diastolic to .astype(float)
data['Systolic'] = data['Systolic'].astype(float)
data['Diastolic'] = data['Diastolic'].astype(float)

#  drop original Blood_Pressure column
data.drop(columns=['Blood_Pressure'], inplace=True)
```

```python
print("SUCCESS CHECKS:")
print("New columns exist : ", 'Systolic' in data.columns and 'Diastolic' in data.columns)
print("\n Systolics stats:")
print(data['Systolic'].describe())
print("\n Diastolics stats:")
print(data['Diastolic'].describe())
print("\n Blood_Pressure gone:",'Blood_Pressure' not in data.columns)
print("\n First 3 rows :")
print(data[['Systolic','Diastolic','Patient_ID']].head(3))
```

```python
# check BMI outliers
print("\n BMI before clipping:",data['BMI'].describe())
data['BMI'] = data['BMI'].clip(lower=15, upper=50)
print("\n BMI after clipping:",data['BMI'].describe())
```

```python
# Data types and missing values after cleaning in medication and data_visited
data['Visit_Date'] = pd.to_datetime(data['Visit_Date'])
# print(data['Visit_Date'].dtype)
# Replacing missing values in Medication with 'None'
data['Medication']=data['Medication'].fillna('None')
# print(data['Medication'].isnull().sum())
```

```python
print("FINAL DATA QUALITY CHECK:")
print("Shape:", data.shape)
print("\nAll nulls gone?", data.isnull().sum().sum() == 0)
print("\nData types perfect:")
print(data.dtypes)
print("\nFirst 3 rows (cleaned):")
print(data[['Patient_ID', 'Systolic', 'Diastolic', 'BMI', 'Visit_Date', 'Medication']].head(3))
```

```python
data.to_csv("..\\data\\cleaned_medical_data.csv", index=False)
print("\nCleaned data saved to ..\\data\\cleaned_medical_data.csv")
print(f"Final shape:{data.shape}")
```
